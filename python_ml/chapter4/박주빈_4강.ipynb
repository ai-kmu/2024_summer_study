## 4.5 편미분

편미분에 의해 임의의 (w0, w1)에서 두 방향의 기울기를 계산할 수 있음.

이것을 f의 w에 대한 경사(기울기 벡터, gradient)라고 부름.

2024-07-07 15;03;23.PNG

### 4.5.3 경사를 그림으로 나타내기

```python
import numpy as np
import matplotlib.pyplot as plt

def f(w0, w1): # f의 정의
  return w0**2 + 2*w0*w1 + 3
def df_dw0(w0, w1): # f의 w0에 관한 편미분
  return 2 * w0 + 2 * w1
def df_dw1(w0, w1): # f의 w1에 관한 편미분
  return 2 * w0 + 0 * w1

w_range = 2
dw = 0.25

w0 = np.arange(-w_range, w_range + dw, dw)
w1 = np.arange(-w_range, w_range + dw, dw)
wn = w0.shape[0]
ww0, ww1 = np.meshgrid(w0, w1)
ff = np.zeros((len(w0), len(w1)))
dff_dw0 = np.zeros((len(w0), len(w1)))
dff_dw1 = np.zeros((len(w0), len(w1)))
for i0 in range(wn):
  for i1 in range(wn):
    ff[i0, i1] = f(ww0[i0, i1], ww1[i0, i1])
    dff_dw0[i1, i0] = df_dw0(w0[i0], w1[i1])
    dff_dw1[i1, i0] = df_dw1(w0[i0], w1[i1])

# 함수 f의 등고선
plt.figure(figsize=(9, 4))
plt.subplots_adjust(wspace=0.3)
plt.subplot(1, 2, 1)
cont = plt.contour(ww0, ww1, ff, 10, color='k')
cont.clabel(fmt='%2.0f', fontsize=8)
plt.xticks(range(-w_range, w_range + 1, 1))
plt.yticks(range(-w_range, w_range + 1, 1))
plt.xlim(-w_range - .5, w_range + .5)
plt.ylim(-w_range - .5, w_range + .5)
plt.xlabel(r'$w_0$', fontsize=14)
plt.ylabel(r'$w_1$', fontsize=14)

# 함수 f의 gradient를 벡터로 표현
plt.subplot(1, 2, 2)
plt.quiver(ww0, ww1, dff_dw0, dff_dw1)
plt.xlabel(r'$w_0$', fontsize=14)
plt.ylabel(r'$w_1$', fontsize=14)
plt.xticks(range(-w_range, w_range + 1, 1))
plt.yticks(range(-w_range, w_range + 1, 1))
plt.xlim(-w_range - .5, w_range + .5)
plt.ylim(-w_range - .5, w_range + .5)
plt.show()
```

오른쪽 그림의 화살표를 따라가면, 어느 지점에서 시작하더라도 그래프의 보다 높은 부분으로 진행한다. 반대로 화살표를 거꾸로 더음어보면 지형의 보다 낮은 부분으로 진행한다.

이렇게 경사는 그 함수의 최대점과 최소점을 찾는 데 중요한 개념이다. 머신러닝에서는 오차 함수의 최소점을 구하기 위해 오차 함수의 경사를 계산한다.

### 4.7.6 소프트맥스 함수

예를 들어, 3개의 수 x0 = 2, x1 = 1, x2 = -1가 있고, 이 수의 대소 관계를 유지하면서, 각각의 확률을 나타내는 y0, y1, y2로 변환하려고 한다.

확률이므로 0에서 1 사이의 숫자를 가지며 모두 더하면 1이어야 한다. 이러한 경우에 사용되는 것이 소프트맥스 함수이다.
2024-07-07 18;11;51.PNG
2024-07-07 18;12;08.PNG

```python
def softmax(x0, x1, x2):
  u = np.exp(x0) + np.exp(x1) + np.exp(x2)
  return np.exp(x0) / u, np.exp(x1) / u, np.exp(x2) / u

y = softmax(2, 1, -1)
print(np.round(y, 2))
print(np.sum(y))
```

```python
from mpl_toolkits.mplot3d import Axes3D

xn = 20
x0 = np.linspace(-4, 4, xn)
x1 = np.linspace(-4, 4, xn)

y = np.zeros((xn, xn, 3))
for i0 in range(xn):
  for i1 in range(xn):
    y[i1, i0, : ] = softmax(x0[i0], x1[i1], 1)

xx0, xx1 = np.meshgrid(x0, x1)
plt.figure(figsize=(8, 3))
for i in range(2):
  ax = plt.subplot(1, 2, i + 1, projection='3d')
  ax.plot_surface(xx0, xx1, y[:, :, i], rstride=1, cstride=1, alpha=0.3, color='blue', edgecolor='black')
  ax.set_xlabel('$x_0$', fontsize=14)
  ax.set_ylabel('$x_1$', fontsize=14)
  ax.view_init(40, -125)

plt.show()
```

### 4.7.9 2차원 가우스 함수

먼저 함수의 형태를 나나태는 매개 변수는 µ와 Σ이다. µ는 평균 벡터로 불리는 매개 변수이며, 분포의 중심을 나타낸다.

2024-07-07 18;37;43.PNG

Σ는 공분산 행렬로, 2*2 행렬이다.

2024-07-07 18;37;53.PNG

행렬 요소의 σ0와 σ1, x0 방향과 x1 방향의 분포의 퍼짐을 나타낸다. σ2는 분포의 기울기(x0과 x1의 상관관계)에 대응된다. 양수라면 오른쪽으로 치우친 분포가, 음수를 넣으면 왼쪽으로 치우친 분포가 표현된다.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline

def gauss(x, mu, sigma):
  N, D = x.shape
  c1 = 1 / (2 * np.pi) ** (D / 2)
  c2 = 1 / (np.linalg.det(sigma) ** (1 / 2))
  inv_sigma = np.linalg.inv(sigma)
  c3 = x - mu
  c4 = np.dot(c3, inv_sigma)
  c5 = np.zeros(N)
  for d in range(D):
    c5 = c5 + c4[:, d] * c3[:, d]
  p = c1 * c2 * np.exp(-c5 / 2)
  return p

x = np.array([[1, 2], [2, 1], [3, 4]])
mu = np.array([1, 2])
sigma = np.array([[1, 0], [0, 1]])
print(gauss(x, mu, sigma))
```

```python
X_range0 = [-3, 3]
X_range1 = [-3, 3]

# 등고선 표시
def show_contour_gauss(mu, sig):
    xn = 40
    x0 = np.linspace(X_range0[0], X_range0[1], xn)
    x1 = np.linspace(X_range1[0], X_range1[1], xn)
    xx0, xx1 = np.meshgrid(x0, x1)
    x = np.c_[np.reshape(xx0, xn * xn), np.reshape(xx1, xn * xn)]
    f = gauss(x, mu, sig)
    f = f.reshape(xn, xn)
    f = f.T
    cont = plt.contour(xx0, xx1, f, 15, colors='k')
    plt.grid(True)

# 3D 표시
def show3d_gauss(ax, mu, sig):
    xn = 40
    x0 = np.linspace(X_range0[0], X_range0[1], xn)
    x1 = np.linspace(X_range1[0], X_range1[1], xn)
    xx0, xx1 = np.meshgrid(x0, x1)
    x = np.c_[np.reshape(xx0, xn * xn), np.reshape(xx1, xn * xn)]
    f = gauss(x, mu, sig)
    f = f.reshape(xn, xn)
    f = f.T
    ax.plot_surface(xx0, xx1, f, rstride=2, cstride=2, alpha=0.3, color='blue', edgecolor='black')

# 메인
mu = np.array([1, 0.5])
sigma = np.array([[2, 1], [1, 1]])
Fig = plt.figure(1, figsize=(7, 3))
Fig.add_subplot(1, 2, 1)
show_contour_gauss(mu, sigma)
plt.xlim(X_range0)
plt.ylim(X_range1)
plt.xlabel('$x0$', fontsize=14)
plt.ylabel('$x1$', fontsize=14)
Ax = Fig.add_subplot(1, 2, 2, projection='3d')
show3d_gauss(Ax, mu, sigma)
Ax.set_zticks([0.05, 0.10])
Ax.set_xlabel('$x0$', fontsize=14)
Ax.set_ylabel('$x1$', fontsize=14)
Ax.view_init(40, -100)
plt.show()
```
