## 4.4 미분

머신러닝은 결국 함수에서 최소나 최대인 입력을 찾는 문제(최적화 문제)다. 함수의 최소지점은 기울기가 0이 되는 성질이 있으므로 이러한 문제를 풀려면 함수의 기울기를 아는 것이 중요하다. 그 함수의 기울기를 도출하는 방법이 '미분'이다.

### 4.4.1 다항식의 미분

$f(w) = w^2$ \\

$\frac{dw}{df(w)},\quad \frac{d}{dw}f(w), \quad f^\prime(w)$ \\

$\frac{d}{dw}w^2 = 2w$ \\

$\frac{d}{dw}w^n = nw^{n-1}$ \\




### 4.4.2 미분 기호가 들어간 수식의 변형

$\frac{d}{dw}2w^5 = 2\frac{d}{dw}w^5 = 2 \times 5w^4 = 10w^4$ \\

$\frac{d}{dw}3 = 0$ \\

$f(w) = a^3 + xb^2 + 2, \quad \frac{d}{dw}f(w) = \frac{d}{dw}(a^3 + xb^2 + 2) = 0$ \\

$f(w) = 2w^3 + 3w^2 + 2, \quad \frac{d}{dw}f(w) = 2\frac{d}{dw}w^3 + 3\frac{d}{dw}w^2 + 2\frac{d}{dw} = 6w^2 + 6w$ \\

### 4.4.3 중첩된 함수의 미분

$f(w) = g(w)^2$ \\

$g(w) = aw + b$ \\

$f(w) = (aw + b)^2 = a^2w^2 + 2abw + b^2$ \\
$\frac{d}{dw}f(w) = wa^2w + 2ab$ \\

### 4.4.4 중첩된 함수의 미분: 연쇄 법칙

$\frac{df}{dw} = \frac{df}{dg}\cdot\frac{dg}{dw}$ \\

$f(w) = g(w)^2$ \\

$g(w) = aw + b$ \\

$\frac{df}{dg} = \frac{d}{dg}g^2 = 2g$ \\

$\frac{df}{dw} = \frac{df}{dg}\cdot\frac{dg}{dw} = 2ga = 2(aw + b)a = 2a^2w + 2ab$ \\

$f(w) = f(g(h(w)))$ \\

$\frac{df}{dw} = \frac{df}{dg}\cdot\frac{dg}{dh}\cdot\frac{dh}{dw}$ \\

## 4.5 편미분

### 4.5.1 편미분이란

$f(w_0, w_1) = w_0^2 + 2w_0w_1 + 3$ \\

$\frac{\partial f}{\partial w_0}, \quad \frac{\partial}{\partial w_0}f, \quad f_{w_0}^\prime$ \\

$f(w_0, w_1) = w_0^2 + 2w_1 \cdot w_0 + 3$ \\

$\frac{\partial f}{\partial w_0} = 2w_0 + 2w_1$ \\

$f(w_0, w_1) = 2w_0 \cdot w_1 + w_0^2 + 3$ \\

$\frac{\partial f}{\partial w_1} = 2w_0$ \\

### 4.5.2 편미분과 도형

image.png

$w_0$와 $w1$에 대한 편미분은 각각 $w_0$방향의 기울기, $w_1$방향의 기울기를 부여할 것이다. 편미분에 의해 임의의 $(w_0, \, w_1)$에서 두 방향의 기울기를 계산할 수 있는 것이다. \\

이 두 개의 기울기를 세트로 하여 벡터로 해석할 수 있다. 이것을 $f$의 **w**에 대한 경사(기울기 벡터, gradient)라고 부르며, 다음과 같이 나타낸다. 경사는 기울기가 가장 큰 방향과 그 크기를 나타낸다.

$\nabla_wf = \begin{bmatrix}\frac{\partial f}{\partial w_0}\\\frac{\partial f}{\partial w_1}\\ \end{bmatrix}$ \\

### 4.5.3 경사를 그림으로 나타내기

$f$를 등고선으로 표시하고, **w**의 공간을 격자형으로 나눴을 때 각 점에서의 경사 $\nabla_w f$를 화살표로 그린다.

```python
import numpy as np
import matplotlib.pyplot as plt

def f(w0, w1): # (A) f의 정의
  return w0**2 + 2 * w0 * w1 + 3
def df_dw0(w0, w1): # (B) f의 w0에 관한 편미분
  return 2 * w0 + 2 * w1
def df_dw1(w0, w1): # (C) f의 w1에 관한 편미분
  return 2 * w0 + 0 * w1

w_range = 2
dw = 0.25

w0 = np.arange(-w_range, w_range + dw, dw)
w1 = np.arange(-w_range, w_range + dw, dw)
wn = w0.shape[0]
ww0, ww1 = np.meshgrid(w0, w1) # (D) 격자 모양으로 나눈 w0와 w1을 2차원 배열 ww0과 ww1에 저장
ff = np.zeros((len(w0), len(w1)))
dff_dw0 = np.zeros((len(w0), len(w1)))
dff_dw1 = np.zeros((len(w0), len(w1)))
for i0 in range(wn): # (E) ww0과 ww1에 대한 f와 편미분의 값
  for i1 in range(wn):
    ff[i1, i0] = f(w0[i0], w1[i1])
    dff_dw0[i1, i0] = df_dw0(w0[i0], w1[i1])
    dff_dw1[i1, i0] = df_dw1(w0[i0], w1[i1])

plt.figure(figsize=(9, 4))
plt.subplots_adjust(wspace=0.3)
plt.subplot(1, 2, 1)
cont = plt.contour(ww0, ww1, ff, 10, colors='k') # (F) f의 등고선 표시
cont.clabel(fmt='%2.0f', fontsize=8)
plt.xticks(range(-w_range, w_range + 1, 1))
plt.yticks(range(-w_range, w_range + 1, 1))
plt.xlim(-w_range - 0.5, w_range + .5)
plt.ylim(-w_range - 0.5, w_range + .5)
plt.xlabel('$w_0$', fontsize = 14)
plt.ylabel('$w_1$', fontsize = 14)

plt.subplot(1, 2, 2)
plt.quiver(ww0, ww1, dff_dw0, dff_dw1) # (G) f의 경사 벡터 표시
plt.xlabel('$w_0$', fontsize = 14)
plt.ylabel('$w_1$', fontsize = 14)
plt.xticks(range(-w_range, w_range + 1, 1))
plt.yticks(range(-w_range, w_range + 1, 1))
plt.xlim(-w_range - 0.5, w_range + .5)
plt.ylim(-w_range - 0.5, w_range + .5)
plt.show()
```

image.png

[그림 4-13] 왼쪽의 $f$의 등고선 플롯의 수치를 보면 $f$는 오른쪽 위와 왼쪽 아래가 높고, 왼쪽 위와 오른쪽 아래가 낮은 지형을 볼 수 있다. [그림 4-13] 오른쪽이 이 지형의 경사다. 화살표는 각 점에서 경사가 높은 쪽을 향하고 있다. 또한 경사가 가파른 정도(등고선의 간격이 좁을수록) 화살표가 긴 것도 알 수 있다. \\

화살표를 따라가면, 어느 지점에서 시작하더라도 그래프의 보다 높은 부분으로 진행한다. 반대로 화살표를 거꾸로 더듬어보면 지형의 보다 낮은 부분으로 진행한다. 이렇게 경사는 그 함수의 최대점과 최소점을 찾는 데 중요한 개념이다. 머신러닝에서는 오차 함수의 최소점을 구하기 위해 오차 함수의 경사를 계산한다.

### 4.5.4 다변수의 중첩 함수의 미분

이 문제는 여러 층으로 된 신경망의 학습 규칙을 도출할 때 등장한다. \\

예를 들어, $g_0$와 $𝙜_0$가 $w_0$와 $w_1$의 함수에서, $f$가 $g_0$와 $g_1$의 함수가 된 경우이다. \\

'$f$의 $w_0$에 관한 미분, $w_1$에 대한 미분은 연쇄 법칙을 사용하면 어떻게 나타낼 수 있는가?'

$f(g_0(w_0, w_1), g_1(w_0,w_1))$ \\
image.png

결론부터 말하면, $w_0$에 대한 미분은 다음과 같다. \\
$\frac{\partial}{\partial w_0} f(g_0(w_0,w_1),g_1(w_0,w_1)) = \frac{\partial f}{\partial g_0} \cdot \frac{\partial g_0}{\partial w_0} + \frac{\partial f}{\partial g_1} \cdot \frac{\partial g_1}{\partial w_0}$ \\

$w_1$에 대한 미분은 다음과 같다. \\
$\frac{\partial}{\partial w_0} f(g_0(w_0,w_1),g_1(w_0,w_1)) = \frac{\partial f}{\partial g_0} \cdot \frac{\partial g_0}{\partial w_1} + \frac{\partial f}{\partial g_1} \cdot \frac{\partial g_1}{\partial w_1}$ \\




예를 들어 $f$가 아래와 같이 나타날 경우, $\partial f / \partial w_0$를 구해보자. \\
$f = (g_0 + 2g_1 - 1)^2, \quad g_0 = w_0 + 2w_1 + 1, \quad g_1 = 2w_0 + 3w_1 - 1$ \\

$\frac{\partial f}{\partial g_0} = 2(g_0 + 2g_1 - 1)$ \\
$\frac{\partial f}{\partial g_1} = 2(g_0 + 2g_1 - 1) \cdot 2$ \\
$\frac{\partial g_0}{\partial w_0} = 1$ \\
$\frac{\partial g_1}{\partial w_0} = 2$ \\

$\frac{\partial f}{\partial w_0} = 2(g_0 + 2g_1 - 1) \cdot 1 + 2(g_0 + 2g_1 - 1) \cdot 2 = 10g_0 + 20_1 - 10$ \\

실제로 신경망의 학습 규칙을 가르칠 경우에는 $f(g_0(w_0,w_1), g1(w_0,w_1), .. , g_M(w_0,w_1))$와 같이 2개 이상의 함수가 중첩된다. \\

$\frac{\partial}{\partial w_0} f(g_0(w_0,w_1),g_1(w_0,w_1)) = \frac{\partial f}{\partial g_0} \cdot \frac{\partial g_0}{\partial w_0} + \frac{\partial f}{\partial g_1} \cdot \frac{\partial g_1}{\partial w_0} + \cdots + \frac{\partial f}{\partial g_M} \cdot \frac{\partial g_M}{\partial w_0} = \displaystyle\sum_{m=0}^{M}\frac{\partial f}{\partial g_m} \cdot \frac{\partial g_m}{\partial w_0}$ \\

$\frac{\partial}{\partial w}\displaystyle\sum_{n} f_n(w) = \displaystyle\sum_{n} \frac{\partial}{\partial w} f_n(w)$ \\


$J = \frac{1}{N}\displaystyle\sum_{n=0}^{N-1}(w_0x_n + w_1 - t_n)^2 $ \\

###4.6.6 역행렬

역수를 곱하면 1이 되는 수이다. $a$의 역수는 $1/a$이며, $a^{-1}$로 나타낼 수 있다. \\

$a \times a^{-1} = 1$ \\

이와 유사하게, 행렬에도 역행렬을 생각할 수 있다. \\

**AA**$^{-1}$ = **A**$^{-1}$**A** = $𝐈$

**A**가 $2 \times 2$의 정방 행렬인 경우에, **A** = $\begin{bmatrix} a&b\\c&d \end{bmatrix}$로 하면, 그 역행렬은 다음의 공식으로 구할 수 있다. \\

**A**$^{-1}$ = $\frac{1}{ad-dc}\begin{bmatrix} d&-b\\-c&a \end{bmatrix} $ \\

예를 들어, **A** = $\begin{bmatrix} 1&2\\3&4 \end{bmatrix}$이면 \\
**A**$^{-1}$ = $\frac{1}{1\cdot4-2\cdot3}\begin{bmatrix} 4&-2\\-3&1 \end{bmatrix}$ = $ -\frac{1}{2} \begin{bmatrix} 4&-2\\-3&1 \end{bmatrix} $ = $\begin{bmatrix} -2&1\\1.5&-0.5 \end{bmatrix} $ \\

**AA**$^{-1}$ = $\begin{bmatrix} 1&2\\3&4 \end{bmatrix} \cdot -\frac{1}{2} \begin{bmatrix} 4&-2\\-3&1 \end{bmatrix}$ = $-\frac{1}{2} \begin{bmatrix} -2&0\\0&-2 \end{bmatrix}$ = $\begin{bmatrix} 1&0\\0&1 \end{bmatrix}$ \\

```python
# 파이썬에서는 np.linalg.inv(A)에서 A의 역행렬을 구할 수 있다.
A = np.array([[1, 2], [3, 4]])
invA = np.linalg.inv(A)

print(invA)
```

여기서 주의해야 할 점은 역행렬이 존재하지 않는 행렬도 있다는 것이다. $2 \times 2$의 정방 행렬이라면 $ad - bc = 0$이 되는 행렬은 역행렬이 존재하지 않는다. 분수의 분모가 0이 되어 버리기 때문이다.

###4.6.9 행렬과 사상

벡터의 덧셈과 뺄셈에 도형적인 해석이 있듯이, 행렬에도 도형적인 해석이 존재한다. 행렬은 '벡터를 다른 벡터로 변환하는 규칙'으로 해석할 수 있다. \\
벡터라는 숫자의 조합을 '좌표'라고 해석하면 그것을 '공간에 있는 점'으로도 볼 수 있기 때문에 행렬을 '어떤 점을 다른 점으로 이동시키는 규칙'으로 파악할 수 있다. \\
이처럼 그룹(벡터와 점)에서 그룹(벡터와 점)에 대응 관계를 제공하는 규칙을 사상이라고 하며, 행렬은 선형 사상으로 분류된다. \\

$\begin{bmatrix} 2&-1\\1&1 \end{bmatrix} \begin{bmatrix} x\\y \end{bmatrix}$ \\

위의 식을 전개하면 아래의 식처럼 되기 때문에, 행렬 $\begin{bmatrix} 2&-1\\1&1 \end{bmatrix}$는 점 $\begin{bmatrix} x\\y \end{bmatrix}$를 점 $\begin{bmatrix} 2x-y\\x+y \end{bmatrix}$으로 이동시키는 사상으로 해석할 수 있다. \\

$\begin{bmatrix} 2&-1\\1&1 \end{bmatrix} \begin{bmatrix} x\\y \end{bmatrix} = \begin{bmatrix} 2x-y\\x+y \end{bmatrix}$ \\

예를 들어 벡터 $\begin{bmatrix} 1&0 \end{bmatrix}^T$에 대입하면  $\begin{bmatrix} 2&1 \end{bmatrix}^T$로 변환되므로, '점 $\begin{bmatrix} 1&0 \end{bmatrix}^T$이 행렬에 의해 $\begin{bmatrix} 2&1 \end{bmatrix}^T$로 이동한다'고 말할 수 있다. \\
이렇게 하여 여러 지점에서의 이동을 그림으로 그린 것이 아래의 왼쪽이다. 밖으로 향해 퍼지는 소용돌이처럼 생겼다. \\
오른쪽은  $\begin{bmatrix} 2&-1\\1&1 \end{bmatrix}$의 역행렬에 의한 사상을 그린 것이다. 이것은 중심을 향하는 소용돌이 같지만, 원래 행렬의 사상과 반대 방향으로 이동하는 것을 알 수 있다.


image.png

$\begin{bmatrix} 2&-1\\1&1 \end{bmatrix} \begin{bmatrix} x\\y \end{bmatrix} = \begin{bmatrix} 0\\3 \end{bmatrix}$ \\
위의 식은 "행렬 $\begin{bmatrix} 2&-1\\1&1 \end{bmatrix}$에 의한 사상에서 $\begin{bmatrix} 0\\3 \end{bmatrix}$으로 옮겨지는 점은 어떤 점입니까"라는 문제로 해석할 수 있다. \\
그리고 그 해답이 되는 식은 다음과 같다. \\
$\begin{bmatrix} x\\y \end{bmatrix} = \begin{bmatrix} 2&-1\\1&1 \end{bmatrix}^{-1} \begin{bmatrix} 0\\3 \end{bmatrix} = \begin{bmatrix} 1\\2 \end{bmatrix}$ \\
이동한 점 $\begin{bmatrix} 0\\3 \end{bmatrix}$을 역행렬 $\begin{bmatrix} 2&-1\\1&1 \end{bmatrix}^{-1}$에 의해서 이동하기 전으로 되돌리면 점 $\begin{bmatrix} 1\\2 \end{bmatrix}$이었다고 해석된다.

###4.7.1 지수

지수는 '그 수를 여러 번 곱한다'는 곱셈의 횟수에서 출발한 개념이지만, 자연수뿐만 아니라 0에도, 음수에도, 실수에도 확장할 수 있다. \\

$y = a^x$ \\
$a$를 사용함을 강조한다면, '$a$를 밑으로 하는 지수 함수'라고 말한다. 여리서 밑 $a$는 0 이상의 1이 아닌 숫자다. \\
위의 식을 그래프로 그리면 $a > 1$의 경우는 단조 증가 그래프다. $0 < a < 1$의 경우는 단조 감소 그래프다. 함수의 출력은 항상 양수다. \\

image.png

###4.7.2 로그

'로그 함수'는 지수 함수의 입력과 출력을 거꾸로 한 것이다. 즉, 지수 함수의 역함수다. \\
image.png

$y = a^y$ \\
이 식을 $y =$ 형태로 변형하면 아래 식이 된다. \\
$y = \log_a x$ \\
그래프를 그리면, $y = a^y$의 그래프와 $y = x$선에서 대칭인 것을 알 수 있다. \\
image.png

```python
x = np.linspace(-8, 8, 100)
y = 2**x

x2 = np.linspace(0.001, 8, 100) # np.log(0)는 에러가 되므로 0은 포함하지 않음
y2 = np.log (x2) / np.log (2) # 밑을 2로 한 log를 공식(7)로 계산
plt.figure(figsize=(5,5))
plt.plot(x, y, 'black', linewidth=3)
plt.plot(x2, y2, 'cornflowerblue', linewidth=3)
plt.plot(x, x, 'black', linestyle='--', linewidth=1)
plt.ylim(-8, 8)
plt.xlim(-8, 8)
plt.grid(True)
plt.show()
```

$𝐚$가 밑인 로그는 너무 크거나 작은 수를 다루기 쉬운 크기의 수로 만들어 주는 편리한 함수다. 예를 들어 $100000000 = 10^8$을 $a=10$의 로그로 나타내면 $\log_{10} 10^8 = 8$이 되며, $0.000000001 = 10^{-8}$은 $\log_{10} 10^{-8}$이 된다. \\

밑을 명기하지 않고 $\log x$로 쓰면 밑에 $e$를 사용한 셈이다. \\
머신러닝에서는 매우 큰 수나 작은 수를 처리한다. 프로그램에서 이러한 숫자를 취급하는 경우 오버플로를 일으킬 수도 있다. 이러한 숫자는 로그를 통해 오버플로를 방지할 수 있다. \\

로그는 곱셈을 덧셈으로 변환한다. \\
$\log\displaystyle \prod_{n=1}^Nf(n) = \displaystyle\sum_{n=1}^N \log f(n)$ \\
이 변환으로 계산이 편해질 수 있다. 그런 이유로 가능도(우도)는 곱셈으로 표현되는 확률은 로그를 취한 '로그가능도'로 생각하는 것이 상투적인 수단이 되고 있다. \\

함수 $f(x)$가 있고, $f(x)$을 최소화하는 $x^*$를 구하고 싶은 경우, 이때 로그를 취한 $\log f(x)$도 $x = x^*$일 때 최소화 된다. 로그는 단조 증가 함수이기 때문에 최솟값은 바뀌어도, 최솟값을 취하는 값은 변하지 않는다. 이는 최댓값을 찾을 때도 마찬가지다. $f(x)$의 최댓값을 취하는 $f(x)$의 로그를 취해도 바뀌지 않는다.

```python
x = np.linspace(-4, 4, 100)
y = (x - 1)**2 + 2
logy = np.log(y)

plt.figure(figsize=(4,4))
plt.plot(x, y, 'black', linewidth=3)
plt.plot(x, logy, 'cornflowerblue', linewidth=3)
plt.ylim(-4, 8)
plt.xlim(-4, 4)
plt.grid(True)
plt.show()
```

image.png

이 성질로 인해 $f(x)$을 최소화 하는 $x^*$을 구하려고 할 때, $\log f(x)$을 최소화하는 $x^*$을 구하는 기법이 많이 사용된다. \\
$\log\displaystyle \prod_{n=1}^Nf(n) = \displaystyle\sum_{n=1}^N \log f(n)$ \\
이 식과 같이 $f(x)$이 곱의 형태로 나타날 경우에는 $\log$를 취해 합의 형태로 만들어, 계속 미분하기 쉬운 형태가 되기 때문에 편리하다.

###4.7.3 지수 함수의 미분

$y = a^x$의 $x$에 대한 미분은 아래와 같다. \\
$y^\prime = (a^x)^\prime = a^x\log a$

```python
x = np.linspace(-4, 4, 100)
a = 2 # 이 값을 여러 가지로 바꿔보자
y = a**x
dy = np.log(a) * y

plt.figure(figsize=(4,4))
plt.plot(x, y, 'gray', linestyle='--', linewidth=3)
plt.plot(x, dy, 'black', linewidth=3)
plt.ylim(-1, 8)
plt.xlim(-4, 4)
plt.grid(True)
plt.show()
```

image.png

$a = e$일 경우, $\log e = 1$이 되므로 아래 식과 같다. \\
$y^\prime (x) = (e^x)^\prime = e^x$ \\
즉, $a = e$의 경우에는 미분해도 함수의 형태가 변하지 않는다. \\
그런 이유로 $e$를 밑으로 한 지수 함수를 다양한 곳에서 사용한다. 시그모이드 함수와 소프트맥스 함수, 가우스 함수에서 $e$를 사용한다.


###4.7.4 로그 함수의 미분

로그 함수의 미분은 반비례의 아래 식이 된다. \\
$y^\prime(x) = (\log x)^\prime = \frac{1}{x}$ \\

```python
x = np.linspace(0.0001, 4, 100) # 0 이하로 정의할 수 없는
y = np.log(x)
dy = 1 / x

plt.figure(figsize=(4,4))
plt.plot(x, y, 'gray', linestyle='--', linewidth=3)
plt.plot(x, dy, 'black', linewidth=3)
plt.ylim(-8, 8)
plt.xlim(-1, 4)
plt.grid(True)
plt.show()
```

image.png

5장에서는 $y = (\log (1 - x))^\prime$의 형태의 미분도 나오고 있지만, $z = 1 - x$를 $y = \log z, \quad z = 1 - x$로 두면 연쇄 법칙을 사용하여 아래 식을 이끌어 낸다. \\
$\frac{dy}{dx} = \frac{dy}{dz} \cdot \frac{dz}{dx} = \frac{1}{z} \cdot (-1) = -\frac{1}{1-x}$

###4.7.5 시그모이드 함수

$y = \frac{1}{1+e^{-x}}$ \\
$y = \frac{1}{1+\exp(-x)}$ \\

```python
x = np.linspace(-10, 10, 100)
y = 1 / (1 + np.exp(-x))

plt.figure(figsize=(4,4))
plt.plot(x, y, 'black', linewidth=3)
plt.ylim(-1, 2)
plt.xlim(-10, 10)
plt.grid(True)
plt.show()
```

image.png

시그모이드 함수는 음에서 양의 실수를 0에서 1까지의 사이로 변환하기 때문에 확률을 나타낼 때 자주 사용한다. \\
시그모이드 함수는 6장에서 분류 문제로 등장한다. 또한 7장의 신경망에서 뉴런의 특성을 나타내는 중요한 함수로 등장한다. 6장과 7장에서 시그모이드 함수의 미분을 사용한다. \\
$(\frac{1}{f(x)})^\prime = -\frac{f(x)^\prime}{f(x)^2}$ \\
이 $f(x)$의 미분은 $f^\prime(x) = -\exp(-x)$이다. \\
$y^\prime = (\frac{1}{1 + \exp(-x)})^\prime = \frac{-\exp(-x)}{1 + \exp(-x)^2} = \frac{\exp(-x)}{(1+\exp(-x))^2}$ \\

$y^\prime = \frac{1}{1 + \exp(-x)} \cdot \frac{1+\exp(-x)-1}{1+\exp(-x)}$ \\
$y^\prime = \frac{1}{1 + \exp(-x)} \cdot \{ 1-\frac{1}{1+\exp(-x)} \}$ \\
여기서 $1/(1+\exp(-x))$는 자체였기 때문에, $y$로 바꿔 쓰면 아래 식처럼 깔끔한 현태가 된다. \\
$y^\prime = y(1-y)$


###4.7.6 소프트맥스 함수

예를 들어, 3개의 수 $x_0 = 2, x_1 = 1, x_2 = -1$가 있고, 이 수의 대소 관계를 유지하면서, 각각의 확률을 나타내는 $y_0, y_1, y_2$로 변환하려고 한다. 확률이므로 0에서 1 사이의 숫자가 아니면 안 된다. 또한 모두 더하면 1이어야 한다. \\
이런 경우에 사용되는 것이 소프트맥스 함수다. 먼저 각 $x_i$의 $\exp$의 합계 $u$를 구해 둔다. \\
$u = \exp(x_0) + \exp(x_1) + \exp(x_2)$ \\
$y_0 = \frac {\exp(x_0)}{u}, y_1 = \frac {\exp(x_1)}{u}, y_2 = \frac {\exp(x_2)}{u}$ \\

```python
def softmax(x0, x1, x2):
  u = np.exp(x0) + np.exp(x1) + np.exp(x2)
  return np.exp(x0) / u, np.exp(x1) / u, np.exp(x2) / u

# test
y = softmax(2, 1, -1)
print(np.round(y, 2)) # (A) 소수점 2 자리로 최대한 가깝게 표시
print(np.sum(y)) # (B) 합계를 표시
```

```python
# 소프트맥스 함수는 입력과 출력이 3차원이므로 그대로 그릴 수는 없다.
# 그래서 x_2만 x_2 = 1로 고정하여 다양한 x_0과 x_1을 입력했을 때의 y_0과 y_1를 플롯한다.

from mpl_toolkits.mplot3d import Axes3D

xn = 20
x0 = np.linspace(-4, 4, xn)
x1 = np.linspace(-4, 4, xn)

y = np.zeros((xn, xn, 3))
for i0 in range(xn):
  for i1 in range(xn):
    y[i1, i0, :] = softmax(x0[i0], x1[i1], 1)

xx0, xx1 = np.meshgrid(x0, x1)
plt.figure(figsize=(8, 3))
for i in range(2):
  ax = plt.subplot(1, 2, i+1, projection='3d')
  ax.plot_surface(xx0, xx1, y[:, :, i], rstride=1, cstride=1, alpha=0.3, color='blue', edgecolor='black')
  ax.set_xlabel('$x_0$', fontsize=14)
  ax.set_ylabel('$x_1$', fontsize=14)
  ax.view_init(40, -125)
```

image.png

$x_2$를 1로 고정하여 $x_0$과 $x_1$을 움직이면 $y_0$, $y_1$은 0과 1 사이의 값으로 변화한다(왼쪽 그래프). $x_0$이 커지면 $y_0$은 1에 가까워지고, $x_1$이 커지면 $y_1$이 1에 가까워진다. $y_2$는 그림으로 나타내고 있지는 않지만, $y_2$는 1에서 $y_0$는 1에서 $y_0$과 $y_1$을 뺀 나머지이므로 상상할 수 있다.

소프트맥스 함수는 3개의 변수뿐 아니라 그 이상의 변수에 사용할 수 있다. 변수의 수를 $K$로 하면 아래 식과 같이 나타낸다. \\
$y_i = \frac{\exp(x_i)}{\sum_{j=0}^{K-1}\exp(x_j)}$ \\

먼저 $y_0$을 $x_0$으로 편미분한다. \\
$\frac {\partial y_0}{\partial x_0} = \frac {\partial}{\partial x_0} \frac{\exp(x_0)}{u}$ \\
여기서 주의해야 하는 것은 $u$도 $x_0$의 함수라는 것이다. 그래서 아래 식의 미분 공식을 사용하여 $f(x) = u = \exp(x_0) + \exp(x_1+ + \exp(x_2), g(x) = \exp(x_0)$로 한다. \\

$(\frac{g(x)}{f(x)})^\prime = \frac {g^\prime (x)f(x) - g(x)f^\prime(x)}{f(x)^2} $ \\

$f^\prime(x) = \frac {\partial}{\partial x_0}f(x) = \exp(x_0)$ \\

$g^\prime(x) = \frac {\partial}{\partial x_0}g(x) = \exp(x_0)$ \\

$\frac {\partial y_0}{\partial x_0} = (\frac {g(x)}{f(x)})^\prime = \frac {\exp(x_0)u - \exp(x_0)\exp(x_0)}{u^2}$ \\
$ \quad\;\;\; = \frac {\exp(x_0)}{u}(\frac {u-\exp(x_0)}{u})$ \\
$ \quad\;\;\; = \frac {\exp(x_0)}{u}(\frac {u}{u}-\frac {\exp(x_0)}{u})$ \\
여기서 $y_0 = \exp(x_0)/u$였음을 활용하여 아래 식과 같이 나타낸다. \\

$\frac {\partial y_0}{\partial x_0} = y_0(1-y_0)$ \\
시그모이드 함수의 미분과 동일한 형태가 되었다.

다음으로 $y_0$를 $x_1$로 편미분한다. \\
$\frac {\partial y_0}{\partial x_1} = \frac {\partial}{\partial x_1} \frac{\exp(x_0)}{u}$ \\

여기서도 $f(x) = u = \exp(x_0) + \exp(x_1+ + \exp(x_2), g(x) = \exp(x_0)$으로 아래 식을 사용한다. \\
$(\frac{g(x)}{f(x)})^\prime = \frac {g^\prime (x)f(x) - g(x)f^\prime(x)}{f(x)^2} $ \\

$f^\prime(x) = \frac {\partial}{\partial x_1}f(x) = \exp(x_1)$ \\

$g^\prime(x) = \frac {\partial}{\partial x_1}g(x) = \exp(x_1)$ \\

$\frac {\partial y_0}{\partial x_1} = \frac {g^\prime (x)f(x) - g(x)f^\prime(x)}{f(x)^2} = \frac {-\exp(x_0)\exp(x_1)}{u^2}$ \\
$\quad\;\;\; = - \frac {\exp (x_0)}{u} \cdot \frac {\exp(x_1)}{u}$ \\
여기서 $y_0 = \exp(x_0)/u, y_1 = \exp(x_1)/u$임을 사용하면 아래 식을 얻을 수 있다. \\
$\frac{\partial y_0}{\partial x_1} = -y_0y_1$ \\
$\frac{\partial y_j}{\partial x_i} = y_j(I_{ij} - y_i)$ \\
여기서는 $I_{ij}$는 $i \neq j$일 때 0이 되는 함수다. \\


###4.7.7 소프트맥스 함수와 시그모이드 함수

소프트맥스 함수와 시그모이드 함수는 닮아 있다. 여기에서 그 관계를 생각해보자. 2 변수의 경우 소프트맥스 함수는 아래 식과 같다. \\
$y = \frac {e^{x_0}}{e^{x_0} + e^{x_1}}$ \\
분모 분자에 $e^{-x_0}$을 곱하여 정리하면 $e^a e^{-b} = e^{a-b}$라는 공식을 사용하여 아래 식을 얻을 수 있다. \\
$y = \frac {e^{x_0} e^{-x_0}}{e^{x_0} e^{-x_0} + e^{x_1} e^{-x_0}} = \frac {e^{x_0-x_0}}{e^{x_0-x_0} + e^{x_1-x_0}} = \frac {1}{1 + e^{-(x_0-x_1)}} $ \\
여기서 $x = x_0 - x_1$로 두면 아래 식과 같이 시그모이드 함수가 된다. \\
$y = \frac {1}{1+e^{-x}}$ \\
즉, 두 변수의 소프트맥스 함수의 입력 $x_0$, $x_1$을 그 차이 $x = x_0 - x_1$로 나타낸 것이 시그모이드 함수다. 시그모이드 함수를 다변수로 확장한 것이 소프트맥스 함수라고 할 수 있다.

###4.7.8 가우스 함수

가우스 함수는 아래 식과 같은 함수다. \\
$y = \exp(-x^2)$ \\

image.png \\

위 그림의 왼쪽 검은 선처럼 가우스 함수는 $x=0$을 중심으로 종 모양을 하고 있다. 가우스 함수는 곡선을 근사하는 기저 함수로 5장에서 등장한다. \\
이 함수의 중심(평균) $𝝁$ 로 나타내고, 확산되는 크기(표준 편차)를 $𝝈$, 높이를 $𝒂$로 하여 조절 가능한 형태로 하면 아래 식과 같다. \\
$y = a\exp(-\frac {(x-𝝁)^2}{𝝈^2})$ \\

```python
def gauss(mu, sigma, a):
  return a * np.exp(-(x - mu)**2 / sigma**2)

x = np.linspace(-4, 4, 100)
plt.figure(figsize=(4,4))
plt.plot(x, gauss(0, 1, 1), 'black', linewidth=3)
plt.plot(x, gauss(2, 3, 0.5), 'gray', linewidth=3)
plt.ylim(-.5, 1.5)
plt.xlim(-4, 4)
plt.grid(True)
plt.show()
```

가우스 함수에서 확률 분포를 나타낼 수 있지만, 그 경우에는 $x$에 관한 적분이 1이 되도록 $y = a\exp(-\frac {(x-μ)^2}{𝝈^2})$의 $a$를 아래 식처럼 만는다. \\
$a = \frac {1}{(2π𝜎^2)^{1/2}}$ \\

###4.7.9 2차원 가우스 함수

가우스 함수를 2차원으로 확장할 수 있다. 2차원 가우스 함수는 9장 혼합 가우시안 모델에서 등장한다. \\
입력을 2차원 벡터 **x** = $[x_0,x_1]^T$라고 했을 때, 가우스 함수의 기본형은 아래 식과 같다. \\
$y = \exp\{-(x_0^2 + x_1^2)\}$ \\
그래프로 나타내면 [그림 4-36]과 같이 원점을 중심으로 동심원을 가진 종 모양이 된다. \\
image.png \\

이 기본형으로 중심을 이동시키거나 길고 가늘게 만들기 위해 몇 가지 매개 변수를 더한 형태가 아래 식이다. \\
$y = a \cdot \exp\{-\frac{1}{2}(𝐱-𝛍)^T 𝚺^{-1}(𝐱-𝛍)\}$ \\
𝛍와 𝚺는 함수의 형태를 나타내는 매개 변수다. \\
𝛍는 평균 벡터(중심 벡터)로 불리는 매개 변수이며, 분포의 중심을 나타낸다. \\
$𝛍 = \begin{bmatrix}\mu_0&\mu_1 \end{bmatrix} ^ T$ \\

𝚺는 공분산 행렬로, $2 \times 2$ 행렬이다. \\
$𝚺 = \begin{bmatrix}σ_0&σ_2\\σ_2&σ_1 \end{bmatrix}$ \\
행렬 요소의 $σ_0$와 $σ_1$, $x_0$ 방향과 $x_1$방향의 분포의 퍼짐을 나타낸다. $σ_2$는 분포의 기울기($x_0$과 $x_1$의 상관 관계)에 대응하고 있으며, 양수라면 오른쪽으로 치우친 분포가, 음수를 넣으면 왼쪽으로 치우친 분포가 표현된다. \\

$𝛍 = \begin{bmatrix} \mu_0&\mu_1 \end{bmatrix}^T = \begin{bmatrix} 0&0 \end{bmatrix}^T$라고 했을 때, $(𝐱-𝛍)^T𝚺^{-1}(𝐱-𝛍)$를 계산하면 아래 식과 같이 되며, $\exp$의 내용의 실체는 $x_0$과 $x_1$의 2차식(2차 형식)임을 알 수 있다. \\

$(𝐱-𝛍)^T𝚺^{-1}(𝐱-𝛍) = 𝐱^T 𝚺^{-1} 𝐱$ \\
$= \begin{bmatrix} x_0&x_1 \end{bmatrix} \cdot \frac{1}{σ_0\sigma_1 - \sigma_2^2} \cdot \begin{bmatrix}σ_0&-σ_2\\-σ_2&σ_1 \end{bmatrix} \begin{bmatrix} x_0\\x_1 \end{bmatrix}$ \\
$= \frac{1}{σ_0\sigma_1 - \sigma_2^2}(\sigma_1x_0^2 - 2\sigma_2x_0x_1 + \sigma_0x_1^2)$ \\


$a$는 함수의 크기를 제어하는 매개 변수로도 생각할 수 있지만, 가우스 함수에서 확률 분포를 나타내는 경우에는 아래와 같이 설정한다. \\
$a = \frac{1}{2π}\frac{1}{|𝚺|^{1/2}} $ \\
위처럼 하면 입력 공간에서의 적분 값이 1이고, 함수가 확률 분포를 표현할 수 있다. \\
위의 식은 $|𝚺|$는 $𝚺$의 행렬식으로 불리는 양으로, $2 \times 2$ 행렬의 경우에는 아래의 공식으로 계산된 양이다. \\
$|𝐀| = \begin{vmatrix}a&b\\c&d\\ \end{vmatrix} = ad - cd$ \\
따라서 $|𝚺|$는 아래와 같이 나타낼 수 있다. \\
$|𝚺| = \sigma_0\sigma_1 - \sigma_2^2$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
%matplotlib inline

def gauss(x, mu, sigma):
  N, D = x.shape
  c1 = 1 / (2 * np.pi)**(D / 2)
  c2 = 1 / (np.linalg.det(sigma)**(1 / 2))
  inv_sigma = np.linalg.inv(sigma)
  c3 = x - mu
  c4 = np.dot(c3, inv_sigma)
  c5 = np.zeros(N)
  for d in range(D):
    c5 = c5 + c4[:, d] * c3[:, d]
  p = c1 * c2 * np.exp(-c5 / 2)
  return p

x = np.array([[1, 2], [2, 1], [3,4]])
mu = np.array([1, 2])
sigma = np.array([[1, 0], [0, 1]])
print(gauss(x, mu, sigma))
```

```python
X_range0 = [-3, 3]
X_range1 = [-3, 3]

# 등고선 표시 -----------------------------------------------------
def show_contour_gauss(mu, sig):
  xn = 40 # 등고선 표시 해상도
  x0 = np.linspace(X_range0[0], X_range0[1], xn)
  x1 = np.linspace(X_range1[0], X_range1[1], xn)
  xx0, xx1 = np.meshgrid(x0, x1)
  x = np.c_[np.reshape(xx0, xn * xn, 'F'), np.reshape(xx1, xn * xn, 'F')]
  f = gauss(x, mu, sig)
  f = f.reshape(xn, xn)
  f = f.T
  cont = plt.contour(xx0, xx1, f, 15, colors='k')
  plt.grid(True)

  # 3D 표시 -------------------------------------------------------
def show3d_gauss(ax, mu, sig):
  xn = 40 # 등고선 표시 해상도
  x0 = np.linspace(X_range0[0], X_range0[1], xn)
  x1 = np.linspace(X_range1[0], X_range1[1], xn)
  xx0, xx1 = np.meshgrid(x0, x1)
  x = np.c_[np.reshape(xx0, xn * xn, 'F'), np.reshape(xx1, xn * xn, 'F')]
  f = gauss(x, mu, sig)
  f = f.reshape(xn, xn)
  f = f.T
  ax.plot_surface(xx0, xx1, f, rstride=2, cstride=2, alpha=0.3, color='blue', edgecolor='black')

# 메인 -----------------------------------------------------------
mu = np.array([1, 0.5]) # (A)
sigma = np.array([[2, 1], [1, 1]])
Fig = plt.figure(1, figsize=(7,3))
Fig.add_subplot(1, 2, 1)
show_contour_gauss(mu, sigma)
plt.xlim(X_range0)
plt.ylim(X_range1)
plt.xlabel('$x_0$', fontsize=14)
plt.ylabel('$x_1$', fontsize=14)
Ax = Fig.add_subplot(1, 2, 2, projection='3d')
show3d_gauss(Ax, mu, sigma)
Ax.set_zticks([0.05, 0.10])
Ax.set_xlabel('$x_0$', fontsize=14)
Ax.set_ylabel('$x_1$', fontsize=14)
Ax.view_init(40, -100)
plt.show()
```
