```python
#   ----
np.random.seed(seed = 1) #  
N = 200 #  
K = 3 #  

T = np.zeros((N, 3), dtype = np.uint8)
X = np.zeros((N, 2))

X_range0 = [-3, 3] # X0  
X_range1 = [-3, 3] # X1  

Mu  = np.array([[-.5, -.5], [.5, 1.0], [1, -.5]]) #  
Sig = np.array([[.7, .7], [.8, .3], [.3, .8]]) #  
Pi  = np.array([0.4, 0.8, 1]) #    

for n in range(N) :
  wk = np.random.rand()
  for k in range(K) :
    if wk < Pi[k]:
      T[n, k] = 1
      break
  for k in range(2) :
    X[n, k] = np.random.randn() * Sig[T[n, :] == 1, k] + Mu[T[n, :] == 1, k]

#  7-1-(2)
# ---- 2      
TestRatio = 0.5
X_n_training = int(N * TestRatio)

X_train = X[:X_n_training, :]
X_test  = X[X_n_training:, :]

T_train = T[:X_n_training, :]
T_test  = T[X_n_training:, :]

# ----  'class_data.npz' 
np.savez('class_data.npz', X_train = X_train, T_train = T_train, X_test = X_test, T_test = T_test, X_range0 = X_range0, X_range1 = X_range1)
```

```python
#   ----
def Sigmoid(x) :
  y = 1 / (1 + np.exp(-x))
  return y

#  ----
def FNN(wv, M, K, x) :
  N, D = x.shape #  

  w = wv[:M * (D + 1)] #   
  w = w.reshape(M, (D + 1))

  v = wv[M * (D + 1):] #   
  v = v.reshape((K, M + 1))

  b = np.zeros((N, M + 1)) #    
  z = np.zeros((N, M + 1)) #   

  a = np.zeros((N, K)) #    
  y = np.zeros((N, K)) #   

  for n in range(N) :
    #  
    for m in range(M) :
      b[n, m] = np.dot(w[m, :], np.r_[x[n, :], 1]) # (A)
      z[n, m] = Sigmoid(b[n, m])

    #  
    z[n, M] = 1 # 
    wkz = 0
    for k in range(K) :
      a[n, k] = np.dot(v[k, :], z[n, :])
      wkz = wkz + np.exp(a[n, k])
    for k in range(K) :
      y[n, k] = np.exp(a[n, k]) / wkz
  return y, a, z, b

# test ----
WV = np.ones(15)
M = 2
K = 3
FNN(WV, M, K, X_train[:2, :])
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class FNN(nn.Module):
    def __init__(self):
        super(FNN, self).__init__()
        input_dim = 2
        hidden_dim = 2
        output_dim = 3
        self.fc1 = nn.Linear(input_dim + 1, hidden_dim)  # +1  bias
        self.fc2 = nn.Linear(hidden_dim + 1, output_dim)  # +1 bias

    def forward(self, x):
        """   
        ones = torch.ones(x.shape[0], 1)
        x = torch.cat((x, ones), dim=1)
        """
        # fc1 + sigmoid
        z = torch.sigmoid(self.fc1(x))

        """   
        ones = torch.ones(z.shape[0], 1)
        z = torch.cat((z, ones), dim=1)
        """
        # fc2 + softmax
        y = self.fc2(z)
        return F.softmax(y, dim=1)

```

```python
#     ----
def CE_FNN(wv, M, k, x, t) :
  N, D = x.shape
  y, a, z, b = FNN(wv, M, K, x)
  ce = -np.dot(np.log(y.reshape(-1)), t.reshape(-1)) / N
  return ce

# test ----
WV = np.ones(15)
M = 2
K = 3
CE_FNN(WV, M, K, X_train[:2, :], T_train[:2, :])

1.0986122886681098

#  7-1-(6)
#   ---- !!!!
def dCE_FNN_num(wv, M, K, x, t) :
  epsilon = 0.001
  dwv = np.zeros_like(wv)
  for iwv in range(len(wv)) :
    wv_modified = wv.copy()

    wv_modified[iwv] = wv[iwv] - epsilon
    mse1 = CE_FNN(wv_modified, M, K, x, t)

    wv_modified[iwv] = wv[iwv] + epsilon
    mse2 = CE_FNN(wv_modified, M, K, x, t)

    dwv[iwv] = (mse2 - mse1) / (2 * epsilon)
  return dwv

# dVW  ----
def Show_WV(wv, M) :
  N = wv.shape[0]
  plt.bar(range(1, M * 3 + 1), wv[:M * 3], align = "center", color = 'black')
  plt.bar(range(M * 3 + 1, N + 1), wv[M * 3:], align = "center", color = 'cornflowerblue')
  plt.xticks(range(1, N + 1))
  plt.xlim(0, N + 1)

# test ----
M = 2
K = 3
nWV = M * 3 + K * (M + 1)
np.random.seed(1)
WV = np.random.normal(0, 1, nWV)
dWV = dCE_FNN_num(WV, M, K, X_train[:2, :], T_train[:2, :])
print(dWV)

plt.figure(1, figsize = (5, 3))
Show_WV(dWV, M)
plt.show()
```

```python
#  7-1-(11)
#   ----
def dCE_FNN(wv, M, K, x, t) :
  N, D = x.shape

  # wv w v 
  w = wv[:M * (D + 1)]
  w = w.reshape(M, (D + 1))

  v = wv[M * (D + 1):]
  v = v.reshape((K, M + 1))

  # (1) x  y 
  y, a, z, b = FNN(wv, M, K, x)

  #   
  dwv = np.zeros_like(wv)
  dw  = np.zeros((M, D + 1))
  dv  = np.zeros((K, M + 1))

  delta1 = np.zeros(M) # 1 
  delta2 = np.zeros(K) # 2 (k = 0   )

  for n in range(N): # (A)
    # (2)   
    for k in range(K) :
      delta2[k] = (y[n, k] - t[n, k])
    # (3)   
    for j in range(M) :
      delta1[j] = z[n, j] * (1 - z[n, j]) * np.dot(v[:, j], delta2)
    # (4) v  dv 
    for k in range(K) :
      dv[k, :] = dv[k, :] + delta2[k] * z[n, :] / N
    # (4) w  dw 
    for j in range(M) :
      dw[j, :] = dw[j, :] + delta1[j] * np.r_[x[n, :], 1] / N
  # dw dv  dwv 
  dwv = np.c_[dw.reshape((1, M * (D + 1))), dv.reshape((1, K * (M + 1)))]
  dwv = dwv.reshape(-1)
  return dwv

# ----Show VW
def Show_dWV(wv, M):
  N = wv.shape[0]
  plt.bar(range(1, M * 3 + 1), wv[:M * 3], align = "center", color = 'black')

  plt.bar(range(M * 3 + 1, N + 1), wv[M * 3:], align = "center", color = 'cornflowerblue')

  plt.xticks(range(1, N + 1))
  plt.xlim(0, N + 1)

#  
M = 2
K = 3
N = 2
nWv = M * 3 + K * (M + 1)
np.random.seed(1)
WV = np.random.normal(0, 1, nWV)

dWV_ana = dCE_FNN(WV, M, K, X_train[:N, :], T_train[:N, :])
print("analytical dWV")
print(dWV_ana)

dWV_num = dCE_FNN_num(WV, M, K, X_train[:N, :], T_train[:N, :])
print("numerical dWV")
print(dWV_num)

plt.figure(1, figsize = (8, 3))
plt.subplots_adjust(wspace = 0.5)
plt.subplot(1, 2, 1)
Show_dWV(dWV_ana, M)
plt.title('analitical')

plt.subplot(1, 2, 2)
Show_dWV(dWV_num, M)
plt.title('numerical')
plt.show()
```

```python
#  7-2-(1)
import numpy as np
import matplotlib.pyplot as plt
import time
np.random.seed(1) # (A)
import keras.optimizers #(B)
from keras.models import Sequential # (C)
from keras.layers import Dense, Activation #(D) core

#   ----
outfile = np.load('class_data.npz')
X_train = outfile['X_train']
T_train = outfile['T_train']

X_test = outfile['X_test']
T_test = outfile['T_test']

X_range0 = outfile['X_range0']
X_range1 = outfile['X_range1']
```

```python
#  7-2-(2)
#   ----
def Show_data(x, t) :
  wk, n = t.shape
  c = [[0, 0, 0], [.5, .5, .5], [1, 1, 1]]
  for i in range(n) :
    plt.plot(x[t[:, i] == 1, 0], x[t[:, i] == 1, 1], linestyle = 'none', marker = 'o', markeredgecolor = 'black', color = c[i], alpha = 0.8)
  plt.grid(True)
```

```python
#  7-2-(3)
#  
np.random.seed(1)

# ---- Sequential  
model = Sequential()

model.add(Dense(2, input_dim = 2, activation = 'sigmoid', kernel_initializer = 'uniform')) # (A)
model.add(Dense(3, activation = 'softmax', kernel_initializer = 'uniform')) # (B)
sgd = keras.optimizers.SGD(learning_rate = 1, momentum = 0.0, nesterov = False) # (C) #decay 

model.compile(optimizer = sgd, loss = 'categorical_crossentropy', metrics = ['accuracy']) # (D)

# ---- 
startTime = time.time()
history = model.fit(X_train, T_train, epochs = 1000, batch_size = 100, verbose = 0, validation_data = (X_test, T_test)) # (E)

# ---- 
score = model.evaluate(X_test, T_test, verbose = 0) # (F)
print('cross entropy {0:3.2f}, accuracy {1:3.2f}'.format(score[0], score[1]))

calculation_time = time.time() - startTime
print("Calculation time:{0:.3f} sec".format(calculation_time))
```
